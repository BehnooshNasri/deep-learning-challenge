```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'. # Dropped "SPECIAL_CONSIDERATIONS" and "STATUS" for optimization.
application_df = application_df.drop(['EIN', "NAME"], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
  # Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff = 500
app_type = application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = list(app_type[app_type < cutoff].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_type = application_df['CLASSIFICATION'].value_counts()
final_class = class_type[class_type > 1]
final_class
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
class_cutoff = 1000
classifications_to_replace = list(class_type[class_type < class_cutoff].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
n_application_df = pd.get_dummies(application_df)
n_application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = n_application_df.drop("IS_SUCCESSFUL", axis=1).values
y = n_application_df["IS_SUCCESSFUL"].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Getting the input feature and shape
X_train_scaled.shape
```

```python
input_features = X_train_scaled.shape[1]
```

```python
# Create automated neural network using keras-tuner library

def create_model (hp):
  nn_model = tf.keras.models.Sequential()

  # Choosing the activation function
  activation = hp.Choice("activation", ["relu", "tanh", "sigmoid"])

  # Choosing the number of neurons in first layer
  nn_model.add(tf.keras.layers.Dense(units=hp.Int("first_units", min_value=1, max_value=90, step=5), activation=activation, input_dim=input_features))

  # Choosing the number of hidden layers and neurons of hidden layers
  for i in range (hp.Int("num_layers", 1, 5)):
    nn_model.add(tf.keras.layers.Dense(units=hp.Int("units_" + str(1), min_value=1, max_value=30, step=5), activation=activation))

  nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

  # Compile the model
  nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

  return nn_model
```

```python
# Import the keras tuner library
import keras_tuner as kt

# Initialize tuner with random search

tuner = kt.RandomSearch(
    create_model,
    objective="val_accuracy",
    max_trials=50
)
```

```python
# Define early stopping
# Define early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,  # Stop after 10 epochs if no improvement
    mode='min'
)
```

```python
# Run the tuner with early stopping
tuner.search(X_train_scaled, y_train, epochs=50, validation_data=(X_test_scaled, y_test), callbacks=[early_stopping])
```

```python
# Get best model hyperparameters
best_hyper = tuner.get_best_hyperparameters()[0]
best_hyper.values
```

```python

# Evaluate best model against full test data
best_model = tuner.get_best_models()[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
